```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fcmeans import FCM
from sklearn import metrics
from sklearn.metrics import pairwise_distances
from sklearn import datasets
from PGM_Func import MakeDataSet, LoadDataSets


datasets = LoadDataSets(os.getcwd(), '.PGM')
```

```python
#fuzzy_cluster function paramaters - data, number of clusters, fuzzifier

def fuzzy_cluster(data, clusters, fuzzifier):
    data_df = pd.DataFrame(np.rot90(data), columns=["feature1", "feature2"])
    
    model = FCM(n_clusters = clusters, m=fuzzifier) #m is the fuzzifier
    model.fit(np.rot90(data))
    centers = model.centers
    #print("The centers are:  ", centers)
    err = model.error
    #print("The Error is:  ", err)

    labels = model.predict(np.rot90(data))
    #data_df['Cluster FCM'] = labels

    x = data_df['feature1']
    y = data_df['feature2']
    #label = data_df["Cluster FCM"]

    #Plot the data with different color clusters
    fig = plt.figure(figsize=(5,4))
    plt.scatter(x, y, c=labels)

    return data_df, labels

#create dataframe to track silhouette scores for clusters
d = {'Dataset': [], 'Cluster Number': [], 'Silhouette Score': []}
silhouette_scores = pd.DataFrame(data=d)
```

```python
name = ['X1_3', 'X1_4']
dataset_num = 1
num_clusters = 3

for data in name: 
    data, labels = fuzzy_cluster(datasets[dataset_num], num_clusters, 1.7)
    silhouette_scores.loc[len(silhouette_scores.index)] = [dataset_num, num_clusters,metrics.silhouette_score(data, labels, metric='euclidean')]
    num_clusters = num_clusters + 1

silhouette_scores.head()
```
